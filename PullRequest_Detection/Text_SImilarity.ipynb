Due to the fact that in this project we deal text data about codes <br>removing some punctuations and symbols are not necessary like !?<> because in most cases they refer to the code chang so we keep them.

```python
lemmatizer = WordNetLemmatizer()
def process_text(text):
  """The function designed to preprocess text data
  here are the steps
  1. converting the text to lowercase
  removing unnecessary words
  lematizing them"""


  text = str(text).lower()
  list_of_words = [i for i in word_tokenize(text) if i not in [",","''"] and  
                   i.lower() not in stopwords.words("english")]
  lemmetized = " ".join(list(map(lemmatizer.lemmatize,list_of_words)))
  
  return lemmetized
```

```python
data.head()
```

```python
data[['Title', 'Content',	"added_comment"]] = data[['Title', 'Content',	"added_comment"]].applymap(process_text)
```

# TF - IDF

```python
cols  = ["Tilte","Condent","added_comment"]
tf_idf_matrices = {}
for col in cols:
  vectorizer = TfidfVectorizer()
  X = vectorizer.fit_transform(data["Title"].tolist())
  vectorizer.get_feature_names_out()
  df = pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names())
  df["Project"] = data["Project"]
  tf_idf_matrices[col]=df
```
